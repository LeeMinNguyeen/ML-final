```python
from connect_database import connector
import random
```

```python
conn = connector()
conn.connects()
```

```python
datafile = './data/randomemail.txt'
```

```python
def usernamefromemail(email):
    return email.split('@')[0]
```

```python
def generaterandompassword():
    return ''.join(random.choices('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=8))
```

```python
data_list = []
for line in open(datafile):
    email = line.strip()
    username = usernamefromemail(email)
    password = generaterandompassword()
    dataformat = {'email': email, 'username': username, 'password': password}
    data_list.append(dataformat)
data_list
```

```python
conn.ImportData(data_list, "Users")
print("Done")
```
