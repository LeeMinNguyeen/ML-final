```python
from connect_database import connector
import pandas as pd
```

```python
conn = connector()
conn.connects()
```

```python
datafile = 'EURUSD_S1.csv'
```

```python
data = pd.read_csv(f"./data/{datafile}")
data
```

```python
data['Time'] = pd.to_datetime(data['Date'].astype(str) + ' ' + data['Timestamp'])
data = data.drop(columns=['Date', 'Timestamp'])

```

```python
data = data[['Time'] + [col for col in data.columns if col != 'Time']]
data
```

```python
data_list = []
for index, row in data.iterrows():
    data_list.append(row.to_dict())
data_list
```

```python
collection = str(datafile.split('.')[0])
```

```python
conn.createCollection(collection)
```

```python
conn.ImportData(data_list, collection)
print("Done")
```
